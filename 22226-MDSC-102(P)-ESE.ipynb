### 6. Write a program to test the hypothesis 0:=1.75 vs 1:1.75 based on the following observations:
1.83, 1.83, 1.73, 1.82, 1.83, 1.73, 1.99, 1.85, 1.68, 1.87

```python
import numpy as np

vals = np.array([1.83, 1.83, 1.73, 1.82, 1.83, 1.73, 1.99, 1.85, 1.68, 1.87])
mu0 = 1.75
mean = np.mean(vals)
sd = np.std(vals)
print("mean = %.2f standard deviation = %.2f" % (mean,sd ))
if(mean != mu0):
    print("We reject the null hpyothesis")
else:
    print("We do not reject the null hypothesis")
```

### 7. Write a program to plot the following function:
$$(;)=(1+)^,0<<1,>1.$$

```python
import matplotlib.pyplot as plt

f = lambda x, theta: (1+theta)*(x**theta)
theta = np.linspace(-1,100,100)
x = np.linspace(0.85,1,100)

plt.plot(f(x,theta))
```

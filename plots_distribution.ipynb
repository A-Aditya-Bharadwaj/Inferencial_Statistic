#### Binomial

$$
f(x) = {}^nc_xp^x(1-p)^{n-x} , x = 0,1,...,n
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def plot_binom(n,p):
    """
        plot the Binomial function
    """
    x = []
    y = []
    f = lambda x: (factorial(n)/(factorial(n-x) * factorial(x))) * (p**x) * ((1-p)**(n-x))
    for i in range(n+1):
        y.append(i)
        x.append(f(i))
    plt.bar(y,x)
```

```python
plot_binom(30,0.2)
```

### Normal Distribution

$$
    X \hookrightarrow  \mathcal{N}(\mu,\,\sigma^{2})
$$
$$
    f(x) = 1/\sigma\sqrt 2 \pi  exp[-1/2 ((x - \mu)/\sigma)^2]
$$

```python
def Normal_distribution(u ,s):
    x = np.linspace(u-4*s, u+4*s, 100)
    f = lambda x: (1/s*sqrt(2*pi)) * e **((-1/2) * ((x - u)/s) ** 2)
    plt.plot(x,f(x))
    plt.show()
```

```python
Normal_distribution(0,1)
```

### Exponential Distribution

```python
def Exponential(l):
    x = np.linspace(0,50, 100)
    f = lambda x: l * e ** ((-l) * x)
    plt.plot(x,f(x))
    plt.show()
```

```python
Exponential(0.2)
```

### Poisson Distribution

```python
def poisson(l):
    x = []
    y = []
    f = lambda x: (np.exp(-l)) * ((l**x)/factorial(x))
    for i in range(30):
        x.append(i)
        y.append(f(i)) 
    plt.bar(x,y)   
```

```python
poisson(0.1)
```

### Gamma Distribution

```python
from sympy import symbols, integrate, oo
def Gamma(l):
    u = symbols("u", positive =True)
    f = ((u**(l-1)) * (e**(-u)))
    return  integrate(f,(u,0, oo))

def Gamma_Distribution(a, l):  # alpha and lambda
    if(a < 0 or l < 0):
        print("Gamma Distribution is not applicable when alpha or lambda values are negative") 
        return
    
    f = lambda x,a,l : ((l**a)/Gamma(l)) * (x**(a-1)) * (e**(-l * x))
    x = np.linspace(0,10,10)
    plt.plot(x,f(x,a,l))
    
```

```python
Gamma_Distribution(5,10)
```

### Beta Distribution

```python
def Beta(a,b):  # alpha and Beta
    f = lambda x: ((Gamma(a)*Gamma(b))/Gamma(a+b)) * (x**(a-1)) * ((1-x)**(b-1)) 
    x = np.linspace(0,1,100)
    plt.plot(x,f(x))
```

```python
Beta(2,2)
```
